## Get All Info for a Playlist, Store into a File

Lets just get the data we need, and store into dataframes/csvs <br>
When recreating playlists, we just need the name, and the ids for the songs within the playlist

```python
for index, playlist in enumerate(playlists['items']):
    print(
        f"{index} : "
        f"{playlist['name']} | "
        f"{playlist['tracks']['total']} | "
        f"{playlist['id']}"
    )
```

```python
df_list = []
for index, playlist in enumerate(playlists['items']):
    playlist_df = spotify_funcs.get_playlist_tracks(
        spotify,
        playlist
    )
    df_list.append(playlist_df)
```

```python
full_playlists_df = pd.concat(df_list)
```
